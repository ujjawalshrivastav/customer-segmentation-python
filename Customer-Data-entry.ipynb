```python
import pandas as pd
import random

customer_ids = []
ages = []
monthly_spends = []
visits = []

for i in range(1, 51):
    customer_id = f"C{i:03d}"
    age = random.randint(18, 60)
    monthly_spend = int(input(f"Enter Monthly Spend for {customer_id}: "))
    visit = random.randint(1, 15)

    customer_ids.append(customer_id)
    ages.append(age)
    monthly_spends.append(monthly_spend)
    visits.append(visit)

df = pd.DataFrame({
    "CustomerID": customer_ids,
    "Age": ages,
    "MonthlySpend": monthly_spends,
    "VisitsPerMonth": visits
})

df.to_csv("customers.csv", index=False)

print("customers.csv file created successfully!")
```
