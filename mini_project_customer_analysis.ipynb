```python
import pandas as pd 
```

```python
#Load Data 

df = pd.read_csv("Customer.csv")
print(df)
```

```python
print("\nAverage Monthly Spend:", df["MonthlySpend"].mean())
print("Maximum Spend:", df["MonthlySpend"].max())
print("Minimum", df["MonthlySpend"].min())
```

```python
#High value "customers 

high_value = df[df["MonthlySpend"] > 80000]

print("\nHigh Value Customers")
print(high_value)

def segment(spend): 
    if spend >= 100000: 
        return "Premium"
    elif spend >= 60000: 
        return "Standard"
    else:
        return "Low"

df["Segment"] = df["MonthlySpend"].apply(segment)

print("\nCustomer Segments")
print(df)
```
